### 핵심객체 : 데이터프레임(DataFrame)
- R의 데이터프레임을 모방

- 데이터 프레임은 엑셀과 유사
- numpy를 사용하고 있어 계산이 빠름
- 라이브러리 구성
    - 여러 종류의 클래스와 다양한 함수로 구성
    - 시리즈와 데이터 프레임의 자료 구조 제공
    - 시리즈(컬럼이 하나), 데이터프레임(컬럼이 여러 개)
